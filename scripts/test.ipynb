```python
import json

def count_string_chars(data):
    total_len = 0
    if isinstance(data, dict):
        for k, v in data.items():
            # JSONのキー（文字列）もカウントに含める場合は total_len += len(k) を追加します
            total_len += count_string_chars(v)
    elif isinstance(data, list):
        for item in data:
            total_len += count_string_chars(item)
    elif isinstance(data, str):
        total_len += len(data)
    return total_len

total_characters = 0
with open('../logs/simulations/sim_20260305_084156.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        data = json.loads(line)
        total_characters += count_string_chars(data)

print(f"文字型の文字数: {total_characters}")
```
